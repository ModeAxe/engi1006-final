```python
charts = []
dates = ["2019-04-28", "2019-05-28", "2019-06-28", "2019-07-28", "2019-08-28", "2019-09-28","2019-10-28", "2019-11-28", "2019-12-28", "2020-01-28", "2020-02-28", "2020-03-28"]
for i in dates:
    charts.append(bd.ChartData("hot-100", i))
    
```

```python
charts
```

```python
uniqueSongs = []
```

```python
for i in range(len(charts)):
    for j in charts[i]:
        uniqueSongs.append(j)
```

```python
len(uniqueSongs)
```

```python
allLyrics = []
uniqueLyrics = set()
```

```python
song = fetch.search_song(uniqueSongs[0].title, uniqueSongs[0].artist)
```

```python
song.lyrics
```

```python
for i in uniqueSongs:
    song = fetch.search_song(i.title, i.artist)
    time.sleep(2.5)
    if(song != None):        
        lyrics = song.lyrics
        time.sleep(2.5)
        allLyrics.append(lyrics)
    
```
